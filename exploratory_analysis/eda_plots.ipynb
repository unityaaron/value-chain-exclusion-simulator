# Data Management
*At the initial stage of the process, all Farmers' Data are imported and Data Type Printed for accurate calculations*  

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#reading the excel data file
df = pd.read_excel("../jpil_farmers_dataset.xlsx", header=1)
df['Training_Encoded'] = df['Training Attended'].map({'Yes': 1, 'No': 0})

#printing the data types
print(df.dtypes)
```

# Statistical Analysis: Impact of Training
I want to determine if the 16.5% yield improvement observed in trained farmers is consistent. I have used a boxplot to compare the distribution of yield changes between trained and untrained groups. 
- **Observation:** The median yield change for trained farmers is clearly higher.
- **Hypothesis:** Structured training is a statistically significant driver of yield improvement.

```python
print(df.groupby('Training Attended')['Yield Change (%)'].mean())
sns.boxplot(x='Training Attended', y='Yield Change (%)', data=df)
plt.title('Impact of Training on Yield Change (%)')
plt.show()
```

*Exploratory analysis of the target variable reveals a stark bimodal distribution rather than a standard Gaussian bell curve. This clear bifurcation in the outcome space shows a complete structural separation between the treatment and control groups. The training intervention didn't just marginally improve yields—it structurally shifted the trained cohort into an entirely higher production frontier.*

```python
sns.histplot(df['Yield Change (%)'], kde=True)


plt.title('Checking the Shape of Yield Change (%)')
plt.xlabel('Yield Change Percentage')
plt.ylabel('Count of Farmers')


plt.show()
```
